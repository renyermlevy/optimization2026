##  Gradient Descent with Backtracking Line Search:

Given function:
\begin{align*}
    f(x_1, x_2, x_3) = x_{3} \log \Big( e^{\frac{x_{1}} {x_{3}}}+ e^{\frac{x_{2}} {x_{3}}} \Big) + (x_{3}-2)^2 + e^{\frac{1}{x_{1} + x_{2}}}
\end{align*}

$ \textbf{dom} \; f: \{ \mathbf{x} \in \mathbb{R}^3 : x_1 +x _2 >0, x_3 > 0 \}  $

Define the function in the cell below:

```julia
# Função objetivo f(x₁, x₂, x₃)
function f(x)
    x1, x2, x3 = x[1], x[2], x[3]
    
    # Verificar se está no domínio
    if x1 + x2 <= 0 || x3 <= 0
        return Inf  # Fora do domínio -> retorna infinito
    end
    
    termo1 = x3 * log(exp(x1/x3) + exp(x2/x3))
    termo2 = (x3 - 2)^2
    termo3 = exp(1/(x1 + x2))
    
    return termo1 + termo2 + termo3
end
```

Defining the first derivative: 

$\nabla f = [ \partial f/\partial x_1 \; \partial f/\partial x_2 \; \partial f/\partial x_3]^T   $

$$ \implies \nabla f = \begin{Bmatrix}
\frac{e^{\frac{x_{1}} {x_{3}}}}{e^{\frac{x_{1}} {x_{3}}}+ e^{\frac{x_{2}} {x_{3}}}} - \frac{e^{ \frac{1}{x_1 + x_2}}}{(x_1 +x_2)^2}  \\ \\
\frac{e^{\frac{x_{2}} {x_{3}}}}{e^{\frac{x_{1}} {x_{3}}}+ e^{\frac{x_{2}} {x_{3}}}} - \frac{e^{ \frac{1}{x_1 + x_2}}}{(x_1 +x_2)^2} \\ \\
 log(e^{\frac{x_{1}} {x_{3}}}+ e^{\frac{x_{2}} {x_{3}}}) - \frac{x_1 e^{\frac{x_{1}} {x_{3}}} + x_2 e^{\frac{x_{2}} {x_{3}}}}{x_3 ( e^{\frac{x_{1}} {x_{3}}}+ e^{\frac{x_{2}} {x_{3}}}) } + 2(x_3-2)
\end{Bmatrix}$$

#### implement the first derivative 

```julia
# Gradiente da função f(x)
function grad_f(x)
    x1, x2, x3 = x[1], x[2], x[3]
    
    # Verificar domínio
    if x1 + x2 <= 0 || x3 <= 0
        return [NaN, NaN, NaN]
    end
    
    # Denominador comum para os termos com exponenciais
    den = exp(x1/x3) + exp(x2/x3)
    
    # Derivada parcial em relação a x1
    df_dx1 = (exp(x1/x3) / den) - exp(1/(x1 + x2)) / (x1 + x2)^2
    
    # Derivada parcial em relação a x2
    df_dx2 = (exp(x2/x3) / den) - exp(1/(x1 + x2)) / (x1 + x2)^2
    
    # Derivada parcial em relação a x3
    df_dx3 = log(den) - (x1*exp(x1/x3) + x2*exp(x2/x3)) / (x3 * den) + 2*(x3 - 2)
    
    return [df_dx1, df_dx2, df_dx3]
end
```

$

\end{array}

### Algorithm: Gradient Descent

1. **Input:** Starting point $x$ in $\text{dom} \, f$

2. **Repeat until stopping criterion is satisfied:**

    a. $\Delta x := -\nabla f(x)$
    
    b. **Line search:** Choose step size $t$ via backtracking line search
    
    c. **Update:** $x := x + t \Delta x$

```julia
# Função de backtracking line search (CORRIGIDA)
function backtracking(f, grad_func, x, Δx, α, β)
    t = 1.0
    
    # Garantir domínio
    xt = x + t * Δx
    while xt[1] + xt[2] <= 0 || xt[3] <= 0
        t = β * t
        xt = x + t * Δx
    end
    
    g = grad_func(x)
    while f(x + t * Δx) > f(x) + α * t * dot(g, Δx)
        t = β * t
        # Verificar domínio novamente
        xt = x + t * Δx
        if xt[1] + xt[2] <= 0 || xt[3] <= 0
            continue
        end
    end
    
    return t
end

println("Função de backtracking definida!")
```

```julia
# Gradient Descent com backtracking line search
function gradient_descent(f, grad, x0, α, β, ϵ, max_iter=1000)
    x = copy(x0)
    f_values = [f(x)]
    grad_norms = [norm(grad(x))]
    converged = false
    final_k = 0
    
    println("=== GRADIENT DESCENT ===")
    println("Iteração 0: x = $(round.(x, digits=6)), f(x) = $(round(f(x), digits=6))")
    
    for k in 1:max_iter
        g = grad(x)
        Δx = -g
        g_norm = norm(g)
        
        if g_norm ≤ ϵ
            converged = true
            final_k = k
            break
        end
        
        t = backtracking(f, grad, x, Δx, α, β)
        x = x + t * Δx
        
        push!(f_values, f(x))
        push!(grad_norms, g_norm)
        
        println("Iteração $k: x = $(round.(x, digits=6)), f(x) = $(round(f(x), digits=6)), ||grad|| = $(round(g_norm, digits=8))")
        final_k = k
    end
    
    if converged
        println("\n✅ Convergiu na iteração $final_k")
    else
        println("\n⚠️ Atingiu máximo de iterações ($max_iter)")
    end
    
    return x, f_values, grad_norms, final_k
end

# Executar
x_opt_gd, f_vals_gd, grad_norms_gd, iter_gd = gradient_descent(f, grad_f, [3.0, 4.0, 5.0], 0.4, 0.5, 1e-5)

println("\n=== RESULTADO FINAL ===")
println("x* = $(round.(x_opt_gd, digits=6))")
println("f(x*) = $(round(f(x_opt_gd), digits=6))")
println("Número de iterações: $iter_gd")
```

### Newton's Algorithm

**Input:** (defined earlier)
- Starting point $x \in \text{dom} \, f$
- Tolerance $\varepsilon > 0$

**Repeat:**
1. Compute the Newton step and decrement.
   $
   \Delta x_{nt} := -\nabla^2 f(x)^{-1}\nabla f(x); \quad \lambda^2 := \nabla f(x)^T \nabla^2 f(x)^{-1}\nabla f(x)
   $
2. Stopping criterion. $\textbf{quit}$ if $\lambda^2/2 \leq \varepsilon.$
3. Line search. Choose step size $t$ by backtracking line search; ensuring update $ := x + t \Delta x$ lies in $\mathbf{dom} f$ throughout.
4. Update. $x := x + t\Delta x_{nt}$.

```julia
# Hessiana usando diferenças finitas
function hess_f(x; h=1e-5)
    n = length(x)
    H = zeros(n, n)
    
    for i in 1:n
        ei = zeros(n)
        ei[i] = h
        H[:, i] = (grad_f(x + ei) - grad_f(x - ei)) / (2h)
    end
    
    return 0.5 * (H + H')  # garantir simetria
end

println("Função Hessiana definida!")
```

```julia
# Newton's Method - VERSÃO FINAL CORRIGIDA
x_newton = [3.0, 4.0, 5.0]
α_val = 0.4
β_val = 0.5
ϵ_val = 1e-5
iter_newton = 0

println("=== NEWTON'S METHOD ===")
println("Iteração 0: x = $(round.(x_newton, digits=6)), f(x) = $(round(f(x_newton), digits=6))")

for k in 1:100
    g = grad_f(x_newton)
    g_norm = LinearAlgebra.norm(g)
    iter_newton = k
    
    if g_norm ≤ ϵ_val
        println("\n✅ Convergência atingida na iteração $(k)!")
        break
    end
    
    H = hess_f(x_newton)
    Δx = -H \ g
    t = backtracking(f, grad_f, x_newton, Δx, α_val, β_val)
    x_newton = x_newton + t * Δx
    
    println("Iteração $k: x = $(round.(x_newton, digits=6)), f(x) = $(round(f(x_newton), digits=6)), ||grad|| = $(round(g_norm, digits=8))")
end

println("\n=== RESULTADO FINAL ===")
println("x* = $(round.(x_newton, digits=6))")
println("f(x*) = $(round(f(x_newton), digits=6))")
println("Número de iterações: $iter_newton")
```

### Running the BFGS Algorithm: 

**Given:**
- Starting point $x_0$
- Convergence tolerance $\epsilon > 0$ $(=10^{-5})$
- Starting matrix $H_0$ (taken as Identity matrix)
  
**Initialization:** $k \gets 0$

**While:** $\| \nabla f_k \| > \epsilon$
1. Compute search direction by solving:
   $ p_k = - H_k \nabla f_k $
2. Set $x_{k+1} = x_k + \alpha_k p_k$, where $\alpha_k$ is computed from backtracking line search procedure.
3. Define $s_k = x_{k+1} - x_k$ and $y_k = \nabla f_{k+1} - \nabla f_k$.
4. Compute $H_{k+1}$ using BFGS.
5. $k \gets k + 1$

Finally, as mentioned in the implementation note to set $\rho_k $ as a constant after $y^T_k s_k$ gets smaller some certain $\epsilon$ say $10^{-5}$.\
We have used an $\epsilon = 10^{-4}$ and set the value for  $\rho_k = 10^{4}$.


```julia
# BFGS - Quasi-Newton Method
function bfgs(f, grad, x0, α, β, ϵ, max_iter=100)
    n = length(x0)
    x = copy(x0)
    H = Matrix{Float64}(I, n, n)
    
    g = grad(x)
    g_norm = norm(g)
    f_values = [f(x)]
    grad_norms = [g_norm]
    converged = false
    final_k = 0
    
    println("=== BFGS (QUASI-NEWTON) ===")
    println("Iteração 0: x = $(round.(x, digits=6)), f(x) = $(round(f(x), digits=6))")
    
    for k in 1:max_iter
        if g_norm ≤ ϵ
            converged = true
            final_k = k - 1
            break
        end
        
        p = -H * g
        t = backtracking(f, grad, x, p, α, β)
        
        x_new = x + t * p
        g_new = grad(x_new)
        g_new_norm = norm(g_new)
        
        s = x_new - x
        y = g_new - g
        
        sy = dot(s, y)
        if sy > 1e-12
            ρ = 1.0 / sy
            H = (I - ρ * s * y') * H * (I - ρ * y * s') + ρ * s * s'
        end
        
        x = x_new
        g = g_new
        g_norm = g_new_norm
        
        push!(f_values, f(x))
        push!(grad_norms, g_norm)
        
        println("Iteração $k: x = $(round.(x, digits=6)), f(x) = $(round(f(x), digits=6)), ||grad|| = $(round(g_norm, digits=8))")
        final_k = k
    end
    
    if converged
        println("\n✅ Convergiu na iteração $final_k")
    end
    
    return x, f_values, grad_norms, final_k
end

# Executar
x_opt_bfgs, f_vals_bfgs, grad_norms_bfgs, iter_bfgs = bfgs(f, grad_f, [3.0, 4.0, 5.0], 0.4, 0.5, 1e-5)

println("\n=== RESULTADO FINAL ===")
println("x* = $(round.(x_opt_bfgs, digits=6))")
println("f(x*) = $(round(f(x_opt_bfgs), digits=6))")
println("Número de iterações: $iter_bfgs")
```

```julia
# Modelo LP
model = Model(HiGHS.Optimizer)
set_silent(model)

# Variáveis de estado x(k) para k = 0,1,2,3
@variable(model, xmin[i] <= x[i=1:3, k=0:T] <= xmax[i])

# Variáveis de controle u(k) para k = 0,1,2
@variable(model, umin[i] <= u[i=1:2, k=0:T-1] <= umax[i])

# Variáveis auxiliares para valor absoluto
@variable(model, z1[0:T] >= 0)
@variable(model, z2[0:T] >= 0)

# Estado inicial
@constraint(model, x[:, 0] .== x0)

# Dinâmica do sistema
for k in 0:T-1
    @constraint(model, x[:, k+1] .== A * x[:, k] + B * u[:, k])
end

# Restrições de valor absoluto para y1 = c (x1)
for k in 0:T
    @constraint(model, -z1[k] <= x[1, k])
    @constraint(model, x[1, k] <= z1[k])
end

# Restrições de valor absoluto para y3 = h (x3)
for k in 0:T
    @constraint(model, -z2[k] <= x[3, k])
    @constraint(model, x[3, k] <= z2[k])
end

# Função objetivo: minimizar soma dos desvios absolutos
@objective(model, Min, sum(z1[k] + z2[k] for k in 0:T))

# Resolver
optimize!(model)

println("Status: ", termination_status(model))
println("Valor ótimo: ", round(objective_value(model), digits=6))
```

```julia
# Extrair resultados
X_opt = value.(x)
U_opt = value.(u)

println("\n=== ESTADOS ÓTIMOS ===")
for k in 0:T
    println("k=$k: x = [$(round(X_opt[1,k], digits=4)), $(round(X_opt[2,k], digits=4)), $(round(X_opt[3,k], digits=4))]")
end

println("\n=== CONTROLES ÓTIMOS ===")
for k in 0:T-1
    println("k=$k: u = [$(round(U_opt[1,k], digits=4)), $(round(U_opt[2,k], digits=4))]")
end

println("\n=== VERIFICAÇÃO ===")
println("x(0) = [$(round(X_opt[1,0], digits=4)), $(round(X_opt[2,0], digits=4)), $(round(X_opt[3,0], digits=4))]")
println("x($T) = [$(round(X_opt[1,T], digits=4)), $(round(X_opt[2,T], digits=4)), $(round(X_opt[3,T], digits=4))]")
```
